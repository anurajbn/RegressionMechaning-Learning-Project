```python
import pandas as pd
import numpy as np

df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '')
df['Memory'] = df['Memory'].str.replace('TB', '000')

df["HDD"] = 0
df["SSD"] = 0
df["Hybrid"] = 0
df["Flash_Storage"] = 0

# Extract numeric storage size from the 'Memory' column
df['Storage_Size'] = df['Memory'].str.extract('(\d+)').astype(float)

# Assign storage size to the corresponding storage columns based on storage type in 'Memory'
df.loc[df['Memory'].str.contains('SSD'), 'SSD'] = df['Storage_Size']
df.loc[df['Memory'].str.contains('HDD'), 'HDD'] = df['Storage_Size']
df.loc[df['Memory'].str.contains('Flash Storage'), 'Flash_Storage'] = df['Storage_Size']

# Drop the temporary 'Storage_Size' column
df.drop(columns=['Storage_Size'], inplace=True)

# Set other storage columns to 0 when one storage type is detected in the "Memory" column
df.loc[df['SSD'] > 0, ['HDD', 'Hybrid', 'Flash_Storage']] = 0
df.loc[df['HDD'] > 0, ['SSD', 'Hybrid', 'Flash_Storage']] = 0
df.loc[df['Flash_Storage'] > 0, ['SSD', 'HDD']] = 0
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(x_train, y_train)

y_pred = pipe.predict(x_test)

print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

step2 = RandomForestRegressor()  # Use RandomForestRegressor instead of LinearRegression

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(x_train, y_train)

y_pred = pipe.predict(x_test)

print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
